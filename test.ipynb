```python
import re

def transform_item(a):
    try:
        ki = a.index('(')
        return defs[a[:ki]]['short_name'], list(map(lambda x: x.split('*'), re.findall(r'\((.*?)\)', a[ki:])))
    except ValueError:
        return a, ''

dict(map(transform_item, content))
```
